```python
# convert not available and unknown data to null
df = pd.read_csv('orders.csv',na_values = ['Not Available', 'unknown'])
df.head(20)
# df['Ship Mode']
```

```python
# rename columns because the naming column is not good
# df.rename(columns={'Order Id': 'order_id', 'Order Date':'order_date'})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df.head(2)
```

```python
# add discount column
# df['list_price'] * (df['discount_percent'] / 100)
df['discount'] = df['list_price'] * (df['discount_percent'] / 100)
df.head(5)

# add sales coiumn
# df['list_price'] - df['discount']
df['sale_price'] = df['list_price'] - df['discount']


# add profit column
# df['list_price'] - df['cost_price']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
# date in order_date data type is not a datetime
df.dtypes

# convert order_date object into datetime
df['order_date'] = pd.to_datetime(df['order_date'], format= '%Y-%m-%d')
# df.head(5)
df.dtypes
```

```python
# drop list_price, cost_price, discount_percent because we dont need that we further are analysis that we earlier create three column

df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace = True)
# inplace true means it change the actual dataset 
```

```python
import pyodbc
import pandas as pd
import logging

def create_connection():
    conn_str = (
        "DRIVER={ODBC Driver 17 for SQL Server};"
        "SERVER={localhost_name};"
        "DATABASE={database_name};"
        "Trusted_Connection=yes;"
    )
    return pyodbc.connect(conn_str)

def bulk_insert_orders(csv_file):
    try:
        df = pd.read_csv(csv_file)

        # Fix types
        df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
        df['discount'] = pd.to_numeric(df['discount'], errors='coerce').round(2).clip(upper=99999.99)
        df['sale_price'] = pd.to_numeric(df['sale_price'], errors='coerce').round(2).clip(upper=99999.99)
        df['profit'] = pd.to_numeric(df['profit'], errors='coerce').round(2).clip(upper=99999.99)

        df.dropna(inplace=True)  # Remove rows with any NaNs

        records = df.values.tolist()

        conn = create_connection()
        cursor = conn.cursor()

        insert_query = """
        INSERT INTO df_orders (
            order_id, order_date, ship_mode, segment, country, city, state, postal_code, region,
            category, sub_category, product_id, quantity, discount, sale_price, profit
        )
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """

        cursor.executemany(insert_query, records)
        conn.commit()
        conn.close()
        print(f"✅ Inserted {len(records)} records into df_orders.")
    except Exception as e:
        logging.error(f"❌ Bulk insert failed: {e}")

# Run it
bulk_insert_orders('cleaned_order_data.csv')
```
